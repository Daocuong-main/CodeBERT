```python
import os
from collections import Counter

def find_duplicate_files(root_dir):
    # This dictionary will store the file name as key and its count as value
    files_count = Counter()

    # Walking through the directory and subdirectories
    for dirpath, _, filenames in os.walk(root_dir):
        for file in filenames:
            # Increment the count of the file name in the counter
            files_count[file] += 1

    # Filter out files with more than one occurrence
    duplicates = {file: count for file, count in files_count.items() if count > 1}

    return files_count, duplicates

# Replace 'your_directory_path' with the path of your directory
all_files, duplicate_files = find_duplicate_files('/home/bkcs/BigCloneEval/ijadataset/bcb_reduced')

print("All files and their counts:")
for file, count in all_files.items():
    print(f"{file}: {count}")

print("\nDuplicate files and their counts:")
for file, count in duplicate_files.items():
    print(f"{file}: {count}")

# Count of all unique files
print(f"\nTotal unique files: {len(all_files)}")

# Count of all duplicate files
print(f"Total duplicate files: {len(duplicate_files)}")
```

```python
import os
import pandas as pd

# Generator function to yield all file names in the directory and subdirectories
def get_all_file_names(root_dir):
    for dirpath, _, filenames in os.walk(root_dir):
        for file in filenames:
            yield file

# Function to find files not in the DataFrame using a set for faster lookups
def find_files_not_in_dataframe(file_names_generator, dataframe):
    df_set = set(dataframe['NAME'].values)
    return (file for file in file_names_generator if file not in df_set)

# Path to the directory
directory_path = '/home/bkcs/Downloads/dataset'

# Get all file names generator
all_file_names_generator = get_all_file_names(directory_path)

# Assuming df_FUNCTIONS is your DataFrame and 'NAME' is the column of interest
# You would load your DataFrame here, for example:
# df_FUNCTIONS = pd.read_csv('path_to_your_csv.csv')

# Find files not in the DataFrame using the generator
files_not_in_df_generator = find_files_not_in_dataframe(all_file_names_generator, df_FUNCTIONS)

# Print the result
print("Files not in the 'NAME' column of df_FUNCTIONS:")
for file in files_not_in_df_generator:
    print(file)

# Since we're using a generator, we can't easily get the total count without iterating through it
```

```python
import pandas as pd

# Assuming df_new is your original DataFrame and 'SYNTACTIC_TYPE' is the column of interest

# Function to create a DataFrame with equal distribution of unique values in a specified column
def create_sample_df(df, column, n_samples):
    unique_values = df[column].unique()
    samples_per_value = n_samples // len(unique_values)
    
    # Ensuring that we have enough samples for equal distribution
    if samples_per_value == 0:
        raise ValueError(f"Not enough samples to distribute across {len(unique_values)} unique values.")
    
    sample_dfs = []
    for value in unique_values:
        sample_df = df[df[column] == value].sample(n=samples_per_value, replace=True)
        sample_dfs.append(sample_df)
    
    # Concatenate all the small sample DataFrames into one
    result_df = pd.concat(sample_dfs).sample(frac=1).reset_index(drop=True)
    return result_df.head(n_samples)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

def split_and_export(df, directory_path, target_column, suffix=None):
    """
    Splits the input DataFrame into training, validation, and test sets with stratification.
    Exports each set to a txt file at the specified directory path.

    Parameters:
    df (DataFrame): The input DataFrame to split.
    directory_path (str): The file path where the txt files will be saved.
    target_column (str): The name of the column with the target labels for stratification.
    suffix (str): Optional suffix for the output file names.
    """
    # Convert suffix to string if it is not None
    suffix = f"_{suffix}" if suffix else ""
    
    # Split the dataframe into train and temp dataframes with 60% and 40% of the data respectively
    train, temp = train_test_split(df, test_size=0.4, random_state=42, stratify=df[target_column])

    # Split the temp dataframe into validate and test dataframes with 50% of the data each
    validate, test = train_test_split(temp, test_size=0.5, random_state=42, stratify=temp[target_column])

    # Export the dataframes to txt files
    train.to_csv(f'{directory_path}train{suffix}.txt', sep='\t', index=False, header=False)
    validate.to_csv(f'{directory_path}valid{suffix}.txt', sep='\t', index=False, header=False)
    test.to_csv(f'{directory_path}test{suffix}.txt', sep='\t', index=False, header=False)


```
