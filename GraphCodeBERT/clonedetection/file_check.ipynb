```python
from collections import Counter
import random

def read_answers(filename):
    answers = {}
    with open(filename) as f:
        for line in f:
            line = line.strip()
            idx1, idx2, label = line.split()
            answers[(idx1, idx2)] = int(label)
    return answers

def calculate_ratio(answers):
    label_counts = Counter(answers.values())
    total_count = sum(label_counts.values())
    ratio = {label: count / total_count for label, count in label_counts.items()}
    return ratio

def create_sample_file(original_answers, ratio, sample_size, output_filename):
    labels = list(ratio.keys())
    sample_counts = {label: int(ratio[label] * sample_size) for label in labels}
    
    # Adjust sample_counts to sum up to sample_size
    while sum(sample_counts.values()) != sample_size:
        for label in labels:
            if sum(sample_counts.values()) < sample_size:
                sample_counts[label] += 1
            elif sum(sample_counts.values()) > sample_size:
                sample_counts[label] -= 1
    
    # Create a list of samples based on the adjusted sample_counts
    samples = []
    for label, count in sample_counts.items():
        label_samples = [(idx_pair, label) for idx_pair, label_value in original_answers.items() if label_value == label]
        samples.extend(random.sample(label_samples, count))
    
    # Write the samples to the output file
    with open(output_filename, 'w') as f:
        for idx_pair, label in samples:
            f.write(f'{idx_pair[0]}\t{idx_pair[1]}\t{label}\n')

# Usage
answers = read_answers('dataset/Multiclass/test.txt')
ratio = calculate_ratio(answers)
create_sample_file(answers, ratio, 10000, 'dataset/Multiclass/sample_test.txt')
```
